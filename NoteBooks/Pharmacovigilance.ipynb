```python
train_path = os.path.join(data_path, 'train')
all_images = os.listdir(train_path)

# Ø§Ø®ØªÙŠØ§Ø± 5 ØµÙˆØ± Ø¹Ø´ÙˆØ§Ø¦ÙŠØ©
sample_images = random.sample(all_images, 5)
corrupted_images = []  # Ù„ØªØ®Ø²ÙŠÙ† Ø§Ù„ØµÙˆØ± Ø§Ù„ØªØ§Ù„ÙØ©
```

```python
for img_name in sample_images:
    img_path = os.path.join(train_path, img_name)
    img = cv2.imread(img_path)

    if img is None:
        print(f"âš ï¸ Ø§Ù„ØµÙˆØ±Ø© ØªØ§Ù„ÙØ©: {img_name}")
        corrupted_images.append(img_name)
        continue

    # ØªØ­ÙˆÙŠÙ„ Ù…Ù† BGR Ø¥Ù„Ù‰ RGB Ù„Ù„Ø¹Ø±Ø¶ Ø§Ù„ØµØ­ÙŠØ­
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(6,6))
    plt.imshow(img_rgb)
    plt.title(img_name)
    plt.axis('off')
    plt.show()

# Ù…Ù„Ø®Øµ Ø§Ù„ØµÙˆØ± Ø§Ù„ØªØ§Ù„ÙØ© (Ø¥Ù† ÙˆØ¬Ø¯Øª)
if corrupted_images:
    print("\nğŸ“Œ Ø§Ù„ØµÙˆØ± Ø§Ù„ØªØ§Ù„ÙØ©:")
    for name in corrupted_images:
        print(name)
else:
    print("\nâœ… Ø¬Ù…ÙŠØ¹ Ø§Ù„ØµÙˆØ± Ø§Ù„Ù…Ø¹Ø±ÙˆØ¶Ø© Ø³Ù„ÙŠÙ…Ø©.")
```
